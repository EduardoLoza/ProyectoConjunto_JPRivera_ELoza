```python
from functools import reduce
import numpy as np
import pandas as pd
from pandas import DataFrame
import string

```

```python
##1 Juan Pablo
def prob1(n):
    x = np.zeros(n)
    D1=0
    D2=1
    x[0] = D1
    x[1] = D2
    def llenar_vector(i):
        nonlocal x
        x[i+2] = ((n-1)*x[i-1]) + ((n-1)*x[i-2])
    [llenar_vector(i) for i in range(n-2)]
    return x

r1=prob1(3)  
r1
```

```python
##2 Juan Pablo
variable ="relaciónn  requiere,  para obtener  el  valor  de un  cierto  término,  el conocimiento de los dos anteriores"
r=[x==' ' for x in variable]
r2=sum(r)
r2
```

```python
##3 Juan Pablo
def remove_vowels(x):
    vowels=('a','e','i','o','u','á','é','í','ó','ú')
    try:
        x = ''.join(ch for ch in x if ch not in vowels)
    except:
        pass
    return x

r3=remove_vowels(variable)
r3
```

```python
##4 Juan Pablo
l1=variable.split()
l1
r4=list(filter(lambda x: len(x)<4, l1))
r4
```
